```python
import os
import pandas as pd
import numpy as np
from glob import glob
import random
import shutil
from tqdm import tqdm

# 원래 데이터의 image 경로와 json 경로를 넣어놓기
annot_PATH = '/content/drive/MyDrive/NIA 폐플라스틱 객체 검출 예측/Data/train/annotation'
jpg_PATH = '/content/drive/MyDrive/NIA 폐플라스틱 객체 검출 예측/Data/train/image'

# 각 경로로 들어가야 하기 때문에 list 형태로 dir 가져오기
file_list = os.listdir(PATH)

# 우리가 적재해야하는 경로 확인해주기
dir_PATH = '/content/drive/MyDrive/NIA 폐플라스틱 객체 검출 예측/Sample_data/annotation'
image_dir_path = '/content/drive/MyDrive/NIA 폐플라스틱 객체 검출 예측/Sample_data/image'

# 각 경로를 for문을 돌면서 하나하나 들어가기
for dir in tqdm(file_list):

    # 모든 json  가져오기
    data = glob(annot_PATH + '/' + dir + '/' + '*.json')

    # 모든 json에서 63개의 sample 가져오기
    data_list = random.sample(data, 63)

    # list를 for문을 돌면서 json 하나하나 적재해 주기
    for file in data_list:

        # json 파일에서 dir을 뽑기 위해서 만들어 줌
        file_dir = file.split('/')[-2]

        # 마지막 경로의 이름만 따와서 .json을 .jpg로 변경
        image_dir = file.split('/')[-1].replace('.json','.jpg')

        # 최종 적재할 경로(json)
        final_path = dir_PATH + '/' + file_dir

        # 이미지 경로와 적재할 이미지 경로 지정
        image_path = jpg_PATH + '/' + file_dir + '/' + image_dir
        dir_image_path = image_dir_path + '/' + file_dir

        shutil.copy(file, final_path)
        shutil.copy(image_path, dir_image_path)
```
